## 最简单的 AI Conversation — 3 行代码
<!-- 最简单的 AI 对话 — 3 行代码 -->

```python
from openai import OpenAI

client = OpenAI()
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": "用一句话解释什么是 API"}],
)
print(response.choices[0].message.content)
```

## Multi-turnConversation
<!-- 多轮对话 -->

LLM API 本身是无状态的。要实现多轮对话，你需要把历史消息一起传给 API。

```python
from openai import OpenAI

client = OpenAI()

messages = [
    {"role": "system", "content": "你是一个友好的 AI 助手，擅长用简单的类比解释技术概念。"},
    {"role": "user", "content": "什么是 API？"},
]

response = client.chat.completions.create(model="gpt-4o-mini", messages=messages)
reply = response.choices[0].message.content
print(f"AI: {reply}\n")

messages.append({"role": "assistant", "content": reply})
messages.append({"role": "user", "content": "能用一个生活中的例子来解释吗？"})

response = client.chat.completions.create(model="gpt-4o-mini", messages=messages)
print(f"AI: {response.choices[0].message.content}")
```

## 查看 API 返回的完整结构

了解 response 对象的结构有助于后续开发。

```python
print(f"模型: {response.model}")
print(f"Token 用量: {response.usage}")
print(f"完成原因: {response.choices[0].finish_reason}")
```
