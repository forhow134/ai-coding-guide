# 13.4 成本优化实验

本笔记本演示 Prompt Caching 和 Batch API 的成本优化效果。

**预估成本**: $0（仅演示）

```python
# Prompt Caching 概念演示

LONG_SYSTEM_PROMPT = '你是专业助手...' * 500  # 模拟长提示词

# 不使用缓存：每次都发送完整系统提示词
def without_caching():
    system_tokens = 3000
    user_tokens = 50
    calls = 3
    
    total_tokens = (system_tokens + user_tokens) * calls
    cost = total_tokens * 0.15 / 1_000_000
    
    print(f'不使用缓存：')
    print(f'  总Token: {total_tokens:,}')
    print(f'  成本: ${cost:.4f}')
    return cost

# 使用缓存：第一次写入（125%），后续命中（50%）
def with_caching():
    system_tokens = 3000
    user_tokens = 50
    
    first_call = int(system_tokens * 1.25) + user_tokens
    cached_calls = int(system_tokens * 0.5) + user_tokens
    
    total_tokens = first_call + cached_calls * 2
    cost = total_tokens * 0.15 / 1_000_000
    
    print(f'\n使用缓存：')
    print(f'  第1次: {first_call:,} tokens (写缓存)')
    print(f'  第2-3次: {cached_calls:,} tokens/次 (缓存命中)')
    print(f'  总Token: {total_tokens:,}')
    print(f'  成本: ${cost:.4f}')
    return cost

cost1 = without_caching()
cost2 = with_caching()

print(f'\n节省: ${cost1-cost2:.4f} ({(1-cost2/cost1)*100:.1f}%)')
```
